## Manipulación datos textuales (Limpieza)

Para realizarle limpieza a columna *county* de la base de datos counties, podemos utilizar lo siguiente:

* **lower:**Poner en minúscula todo el texto.
* **replace:**Remplazar ciertos valores por otros.
* **strip:**Eliminar los espacios al principio y al final de la cadena.
* **title:**Poner primera letra de cada palabra en mayúscula.

```python
counties["county"] = (counties["county"]
                            .str.lower()
                            .str.replace("[^a-záéíóúüñ ]","")
                            .str.replace(" +"," ")
                            .str.strip()
                            .str.title()
                           ) 
```

## Series  de tiempo y time stamps

Existe un tipo de datos llamado "timestamp" que se usa para medir registros de tiempo. Es posible pasar strings a este tipo de dato en diferentes formatos y trabajar con este usándolo como indice

```python
airrpm=pd.read_csv('airrpm.txt', header=None, delimiter= '\s+', decimal=",")
airrpm.columns=["Time", "R1", "R2", "R3"]
airrpm.head()
```

La variable "Time" en este momento es un string. Pero podemos transformarlo usando `to_datetime()`. Para obtener más información de qué se podría poner en `format`, pueden revisar este [link](https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior)

```python
airrpm['Time']=pd.to_datetime(airrpm['Time'], format='%b-%y')
airrpm.index=airrpm['Time']
airrpm=airrpm.drop(['Time'], axis=1)
airrpm.head()
```

Definiendo la fecha como índice, podemos extraer información con respecto al año y respecto al mes

```python
airrpm.groupby([airrpm.index.month])['R1'].describe()
```

```python
airrpm.groupby([airrpm.index.year]).sum().head(10)
```

#EJEMPLO MUNICIPIOS EN PANDAS

Otras opciones que se pueden hacer con la librería pandas se mostrarán como ejemplos de la clase de computación estadistica

Cuales variables del conjunto de datos Municipios tienen valores perdidos o missings?

```python
Municipios = pd.read_excel("Municipios.xlsx",dtype={"Irural":np.float64,"Dep":str,
                                                    "Poblacion":np.float64,"Depmun":str})
```

```python
def missings(x):
            a = sum(x.isna())
            b = 100*np.mean(x.isna())
            c = pd.Series({"Missings": a,"%Missings": b})
            return(c)
Municipios.apply(missings,axis=0)
Municipios.apply(missings,axis=1)
Municipios.apply(missings,axis=1).query("Missings > 0")
            
```

Se hace "limpieza" en los nombres de los municipios, es decir,

* Se eliminan caracteres especiales
* Se eliminan espacios al inicio y al final
* Se reemplazan espacios múltiples por sencillos
* Se escriben en letras minúsculas excepto en la inicial de cada palabra

```python
Municipios["Municipio2"] = (Municipios["Municipio"]
                            .str.lower()
                            .str.replace("[^a-záéíóúüñ ]","", regex=True)
                            .str.replace(" +"," ", regex=True)
                            .str.strip()
                            .str.title()
                           ) 
Municipios[["Municipio","Municipio2"]].tail(10)
Municipios["Municipio"] = Municipios["Municipio2"]
Municipios = Municipios.drop(columns="Municipio2")
Municipios.tail(5)
```

Se hace "limpieza" en los nombres de los departamentos, es decir,

* Se eliminan caracteres especiales
* Se eliminan espacios al inicio y al final
* Se reemplazan espacios múltiples por sencillos
* Se escriben en letras minúsculas excepto en la inicial de cada palabra

```python
Municipios["Departamento2"] = (Municipios["Departamento"]
                               .str.lower()
                               .str.replace("[^a-záéíóúüñ ]","", regex=True)
                               .str.replace(" +"," ", regex=True)
                               .str.strip()
                               .str.title()
                              ) 
Municipios[["Departamento","Departamento2"]].head(10)
Municipios["Departamento"] = Municipios["Departamento2"]
Municipios = Municipios.drop(columns="Departamento2")
Municipios.head(5)
```

Conjunto de datos con las variables Departamento, Dep, Depmun, Municipio, Superficie y Poblacion que incluye todos los municipios de los departamentos de Antioquia y Caldas con más de 20 mil habitantes y donde los mismos están ordenados con respecto a Dep y de mayor a menor con respecto a Poblacion.

```python
e1 = (Municipios
      .loc[:,["Departamento","Dep","Depmun","Municipio","Superficie","Poblacion"]] 
      .query("Poblacion > 20000 & Dep in ('05','17')")
      .sort_values(by=["Dep","Poblacion"],ascending=[True,False]).reset_index(drop=True)
     )

e1.info()
e1.head(5)
```

Cuantos y cuales municipios tienen al menos una estación de monitoreo climático?

```python
Estaciones = pd.read_excel("Estaciones.xlsx",dtype={'CODIGO': str, 'Depmun': str, 'ELEV':np.float64})
Estaciones.info()
Estaciones.head(5)
```

```python
conest = pd.DataFrame(Estaciones["Depmun"].unique(),columns=["Depmun"])
conest.info()
```

Conjunto de datos con todos los municipios del país, que incluye una variable que informa el número de estaciones de monitoreo climático con que cuenta cada municipio.

```python
Emon = (Estaciones
        .groupby("Depmun")
        .agg(nem=("CODIGO","count")).reset_index()
       )

e19 = (Municipios
      .merge(Emon,left_on="Depmun",right_on="Depmun",how="left")
      .eval('''nest = nem.fillna(0).astype('int64').values''')
      .drop(columns="nem")
      .sort_values(by=["Dep","Poblacion"],ascending=[True,False]).reset_index(drop=True)
      )


e19.info()
e19.head(5)
```
